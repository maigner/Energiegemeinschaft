```python
from dotenv import load_dotenv
import os
from pathlib import Path

# Load the .env file
dotenv_path = Path("./.env")
load_dotenv(dotenv_path=dotenv_path)
os.getenv("FAKTURA_USER")
```

```python
os.getenv("RC_NUMBER")
```

```python
# timestamp from date-time string
def to_timestamp(date_str):
    from datetime import datetime
    dt = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
    return int(dt.timestamp())


# post data
data = {
    "ecId": os.getenv("RC_NUMBER"),
    "start": to_timestamp("2025-01-01 00:00:00"),
    "end": to_timestamp("2025-01-01 23:59:59"),
    "cps": [
        {
            "meteringPoint": "AT0030000000000000000000000396479"
        }
    ]
}
data
```

```python
# make POST request with BASIC auth
import requests
from requests.auth import HTTPBasicAuth

headers = {
    "X-Tenant": os.getenv("RC_NUMBER"),
    "Content-Type": "application/json"
}

response = requests.post(
    "https://eegfaktura.at/energystore/query/rawdata",
    json=data,
    headers=headers,
    auth=(os.getenv("FAKTURA_USER"), os.getenv("PASSWORD"))
)


if response.status_code == 400:
    print("Bad Request (400): Check your payload or parameters.")
    print(response.text)
elif response.status_code in [200, 201]:
    print("Success:", response.json())
else:
    print(f"Error {response.status_code}: {response.text}")
```
