```python
import os
from dotenv import load_dotenv
from dotenv import dotenv_values

# Load the .env file
CONFIG = dotenv_values(".env")
TOKEN = CONFIG.get("OPENHAB_TOKEN")
URL = CONFIG.get("OPENHAB_URL")
```

```python
import requests

url = URL+'/rest/persistence/items/SolarEdge_Consumption'
params = {
    'starttime': '2024-01-15T00:00:00.000+0200'
}
headers = {
    'accept': 'application/json',
    'Authorization': 'Bearer ' + TOKEN
}

response = requests.get(url, headers=headers, params=params)

# Output the result
print(response.status_code)
print(response.json())
```

```python
result = response.json()
data = result['data']
```

```python
import pandas as pd

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert 'time' from Unix ms to datetime
df['time'] = pd.to_datetime(df['time'], unit='ms')

# Convert 'state' to float
df['state'] = df['state'].astype(float)
```
