```python
import requests
import json

# download data
data = {}


# returns JSON

url = "https://archive-api.open-meteo.com/v1/archive?latitude=47.712596&longitude=13.618029&start_date=2024-01-15&end_date=2025-07-05&hourly=temperature_2m,cloud_cover_low,cloud_cover_mid,cloud_cover_high,snow_depth,snowfall,cloud_cover,weather_code,wind_speed_10m,wind_speed_100m,rain,dew_point_2m,sunshine_duration,relative_humidity_2m&timezone=Europe%2FBerlin"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()  # Automatically parses JSON into a Python dictionary
    print("Downloaded JSON successfully!")
    print(data)
else:
    print(f"Failed to download JSON. Status code: {response.status_code}")
```

```python
hourly_units = data['hourly_units']
hourly_units
```

Load CSV into a Python data structure for processing

```python
metrics = list(hourly_units.keys())
metrics
```

```python
import pandas as pd
```

```python
hourly = data["hourly"]
data_frame_setup = {}

for metric in metrics:
    data_frame_setup[metric] = hourly[metric]
    
```

```python
hourly
```

```python
df = pd.DataFrame(data_frame_setup)

df['time'] = pd.to_datetime(df['time'])

```
